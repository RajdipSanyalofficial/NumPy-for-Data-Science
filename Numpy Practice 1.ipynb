# Indexing & Slicing

```python
# ** 1,2,3 use [] inside[] --> [[1,2,3],1:3]
# ** : --> no [] and :: --> no []
# ** in case of jump then use :: 
#  11 to 14 then can write ::3 , 
# 1 to 3 --> 1:4 (4-1=3)
```

```python
z4=np.arange(1,26).reshape((5,5))
z4
```

```python
z4[:,2] #for one column need all rows element
```

```python
z4[2,] #for one row need all columns element
```

```python
# 789
# 12 13 14
# 17 18 19

z4[1:4,1:4] #no jump that's why colon (:) is used
```

```python
# 7,10 12,15
z4[[1,2],1::3] #jump from 7 to 10 --> 3 jump
```

# Fancy Indexing

```python
#1 4 11 14 21 24
z4[[0,2,4],::3]
```

```python
# 2 3 4 12 13 14 22 23 24
z4[[0,2,4],1:4]
```

```python
# # 1,3 4th col
z4[:,[1,3,4]]
```
